```python
import pandas as pd
```

```python
new_1ET = list(pd.read_csv("GRLCL_1D_ETHCD.tsv", sep="\t")["Peptide"])
new_1 = list(pd.read_csv("GRLCL_1D_HCD.tsv", sep="\t")["Peptide"])
new_2ET = list(pd.read_csv("GRLCL_2D_ETHCD.tsv", sep="\t")["Peptide"])

ori_1ET = list(pd.read_csv("GR-LCL_EThcD_1D_spliced.csv").columns) # a
ori_1 = list(pd.read_csv("GR-LCL_HCD_1D_spliced_peptides.csv").columns) # b
ori_2ET = list(pd.read_csv("GR-LCL_EThcD_2D_spliced.csv").columns) # c
```

```python
def matches(new, ori):
    if len(new)==len(set(new)):
        print("Peptides are unique.")
    else:
        print("Peptides are duplicated, duplicates where removed.")
        new = list(set(new))
    count = 0
    for i in ori:
        if i in new:
            count += 1
            print(i)
    return len(ori), count, str(round(count/len(ori)*100, 3))+"%"

a_tot, a_match, a_per = matches(new_1ET, ori_1ET)
b_tot, b_match, b_per = matches(new_1, ori_1)
c_tot, c_match, c_per = matches(new_2ET, ori_2ET)

print(f"\nGR-LCL 1D ETHCD:\nFrom {a_tot} declared spliced peptides {a_match} are non-canonical. It is", a_per)
print(f"\nGR-LCL 1D HCD:\nFrom {b_tot} declared spliced peptides {b_match} are non-canonical. It is", b_per)
print(f"\nGR-LCL 2D ETHCD:\nFrom {c_tot} declared spliced peptides {c_match} are non-canonical. It is", c_per)
```
