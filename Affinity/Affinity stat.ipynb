```python
import pandas as pd
import seaborn as sea
from statistics import median, mean
import matplotlib.pyplot as plt
from scipy import stats
```

```python
s_non = pd.read_csv("NetMHCpan2023(9-12)in-silico_non.csv")
s_spl = pd.read_csv("NetMHCpan2023(9-12)in-silico_spliced.csv")
v_non = pd.read_csv("NetMHCpan2023(9-12)in-vitro_non.csv")
v_spl = pd.read_csv("NetMHCpan2023(9-12)in-vitro_spliced.csv")

def ext(x):
    x = x[5:507]
    new_header = x.iloc[1].where(x.iloc[1] != "core", x.iloc[0])
    x.columns = new_header 
    y = x["BA_score"]
    y.columns = x.iloc[0].dropna()
    y = y[2:507].apply(pd.to_numeric)
    return y


s_non = ext(s_non)
s_spl = ext(s_spl)
v_non = ext(v_non)
v_spl = ext(v_spl)
```

```python
dfs = {"In silico non-spliced": s_non,
    "In silico spliced": s_spl,
    "In vitro non-spliced": v_non,
    "In vitro spliced": v_spl,
}

for col in s_non.columns:
    plt.figure(figsize=(4, 3))
    for label, df in dfs.items():
        sea.kdeplot(
            data=df,
            x=col,
            label=label,
            fill=False,
            common_norm=False            
        )
    res = stats.wilcoxon(v_spl[col], s_spl[col], alternative='two-sided', zero_method='pratt')
    ks_statistic, ks_p_value = stats.ks_2samp(v_spl[col], s_spl[col])
    print("Spliced in silico vs in vitro:\n\nWilcoxon test:\nstatistic:", float(res[0]), "\nP-value:", round(float(res[1]), 4),
          "\n\nKolmogorov-Smirnov test:\nks-statistic:", float(ks_statistic), "\nP-value:", round(float(ks_p_value), 4),
          "\n\nMedian (in silico):", median(s_spl[col]), "\nMedian (in vitro):", median(v_spl[col]),
          "\nMedian (non-spliced, in silico):", median(s_non[col]), "\nMedian (non-spliced, in vitro):", median(v_non[col]))
    plt.title(col)
    plt.legend()
    plt.tight_layout()
    plt.xlabel("Binding affinity")
    plt.ylabel("")
    plt.show()
```

```python
## Total analysis
tot_data = {"In silico non-spliced": [],
    "In silico spliced": [],
    "In vitro non-spliced": [],
    "In vitro spliced": []}
for i in tot_data.keys():
    tot_data[i] = dfs[i].melt(value_name='All_HLAs')

plt.figure(figsize=(6, 4))
for label, df in tot_data.items():
    sea.kdeplot(
        data=df,
        x="All_HLAs",
        label=label,
        fill=False,
        common_norm=False)

plt.title("Total for all HLA allotypes")
plt.legend()
plt.tight_layout()
plt.xlabel("Binding affinity")
plt.ylabel("")
plt.show()

# Spliced in silico vs in vitro
res = stats.wilcoxon(tot_data["In vitro spliced"]["All_HLAs"], tot_data["In silico spliced"]["All_HLAs"], alternative='two-sided', zero_method='pratt')
ks_statistic, ks_p_value = stats.ks_2samp(tot_data["In vitro spliced"]["All_HLAs"], tot_data["In silico spliced"]["All_HLAs"])
print("Spliced in silico vs in vitro:\n\nWilcoxon test:\nstatistic:", float(res[0]), "\nP-value:", float(res[1]),
      "\n\nKolmogorov-Smirnov test:\nks-statistic:", float(ks_statistic), "\nP-value:", float(ks_p_value),
      "\n\nMedian (in silico):", median(tot_data["In silico spliced"]["All_HLAs"]), "\nMedian (in vitro):", median(tot_data["In vitro spliced"]["All_HLAs"]))

# Non-spliced in silico vs in vitro
res = stats.wilcoxon(tot_data["In vitro non-spliced"]["All_HLAs"], tot_data["In silico non-spliced"]["All_HLAs"], alternative='two-sided', zero_method='pratt')
ks_statistic, ks_p_value = stats.ks_2samp(tot_data["In vitro non-spliced"]["All_HLAs"], tot_data["In silico non-spliced"]["All_HLAs"])
print("\n\n\nNon-spliced in silico vs in vitro:\n\nWilcoxon test:\nstatistic:", float(res[0]), "\nP-value:", float(res[1]),
      "\n\nKolmogorov-Smirnov test:\nks-statistic:", float(ks_statistic), "\nP-value:", float(ks_p_value),
      "\n\nMedian (in silico):", median(tot_data["In silico non-spliced"]["All_HLAs"]), "\nMedian (in vitro):", median(tot_data["In vitro non-spliced"]["All_HLAs"]))

# In vitro spliced vs non-spliced
ks_statistic, ks_p_value = stats.ks_2samp(tot_data["In vitro non-spliced"]["All_HLAs"], tot_data["In vitro spliced"]["All_HLAs"])
print("\n\n\nIn vitro spliced vs non-spliced:",
      "\n\nKolmogorov-Smirnov test:\nks-statistic:", float(ks_statistic), "\nP-value:", float(ks_p_value),
      "\n\nMedian (spliced):", median(tot_data["In vitro spliced"]["All_HLAs"]), "\nMedian (non-spliced):", median(tot_data["In vitro non-spliced"]["All_HLAs"]))

# In silico spliced vs non-spliced
ks_statistic, ks_p_value = stats.ks_2samp(tot_data["In silico non-spliced"]["All_HLAs"], tot_data["In silico spliced"]["All_HLAs"])
print("\n\n\nIn silico spliced vs non-spliced:",
      "\n\nKolmogorov-Smirnov test:\nks-statistic:", float(ks_statistic), "\nP-value:", float(ks_p_value),
      "\n\nMedian (spliced):", median(tot_data["In silico spliced"]["All_HLAs"]), "\nMedian (non-spliced):", median(tot_data["In silico non-spliced"]["All_HLAs"]))
```
